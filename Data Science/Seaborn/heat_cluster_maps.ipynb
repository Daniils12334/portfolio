```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue = "pclass", palette = "YlGnBu")
```

```python
sns.heatmap(titanic.select_dtypes(include='number').corr(), annot=True, cmap="coolwarm")
```

```python
sns.clustermap(iris.drop("species", axis=1), standard_scale=1)
```
